```python
import requests
import constants
```

```python
def matchHistory(region = 'na', name = 'Huan', tag = 'ipynb'):
    
    params = {'region' : region, 'name' : name, 'tag' : tag}
    headers = {'Authorization' : constants.VAL_API_KEY}

    api_url = 'https://api.henrikdev.xyz/valorant/v3/matches/{region}/{name}/{tag}'

    res = requests.get(api_url.format(**params), headers = headers)
    return res
```

```python
res = matchHistory()
```

```python
data = res.json()['data']
```
